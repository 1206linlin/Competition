二分帶轉經緯度 refer to https://tylerastro.medium.com/twd97-to-longitude-latitude-dde820d83405

```python
import math
```

```python
#二度分帶轉經緯度
def twd97_to_lonlat(x, y):
    """
    list
        [longitude, latitude] = [經度, 緯度]
    """

    a = 6378137
    b = 6356752.314245
    long_0 = 121 * math.pi / 180.0
    k0 = 0.9999
    dx = 250000
    dy = 0

    e = math.pow((1-math.pow(b, 2)/math.pow(a,2)), 0.5)

    x -= dx
    y -= dy

    M = y / k0

    mu = M / ( a*(1-math.pow(e, 2)/4 - 3*math.pow(e,4)/64 - 5 * math.pow(e, 6)/256))
    e1 = (1.0 - pow((1   - pow(e, 2)), 0.5)) / (1.0 +math.pow((1.0 -math.pow(e,2)), 0.5))

    j1 = 3*e1/2-27*math.pow(e1,3)/32
    j2 = 21 * math.pow(e1,2)/16 - 55 * math.pow(e1, 4)/32
    j3 = 151 * math.pow(e1, 3)/96
    j4 = 1097 * math.pow(e1, 4)/512

    fp = mu + j1 * math.sin(2*mu) + j2 * math.sin(4* mu) + j3 * math.sin(6*mu) + j4 * math.sin(8* mu)

    e2 = math.pow((e*a/b),2)
    c1 = math.pow(e2*math.cos(fp),2)
    t1 = math.pow(math.tan(fp),2)
    r1 = a * (1-math.pow(e,2)) / math.pow( (1-math.pow(e,2)* math.pow(math.sin(fp),2)), (3/2))
    n1 = a / math.pow((1-math.pow(e,2)*math.pow(math.sin(fp),2)),0.5)
    d = x / (n1*k0)

    q1 = n1* math.tan(fp) / r1
    q2 = math.pow(d,2)/2
    q3 = ( 5 + 3 * t1 + 10 * c1 - 4 * math.pow(c1,2) - 9 * e2 ) * math.pow(d,4)/24
    q4 = (61 + 90 * t1 + 298 * c1 + 45 * math.pow(t1,2) - 3 * math.pow(c1,2) - 252 * e2) * math.pow(d,6)/720
    lat = fp - q1 * (q2 - q3 + q4)


    q5 = d
    q6 = (1+2*t1+c1) * math.pow(d,3) / 6
    q7 = (5 - 2 * c1 + 28 * t1 - 3 * math.pow(c1,2) + 8 * e2 + 24 * math.pow(t1,2)) * math.pow(d,5) / 120
    lon = long_0 + (q5 - q6 + q7) / math.cos(fp)

    lat = (lat*180) / math.pi
    lon = (lon*180) / math.pi
    return lon, lat
```

```python
#計算經緯度距離
from math import radians, cos, sin, asin, sqrt

#address_me = [120.2640171785886,22.831598506259798]

def get_address_distance(station_lon, station_lat, train_lon,  train_lat):
    lon1, lat1, lon2, lat2 = map(radians, [station_lon, station_lat, train_lon, train_lat])
    # 半正矢公式
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    c = 2 * asin(sqrt(a))
    r = 6371
    dis = round((c * r * 1000)/1000, 2)
    return dis

#instance = get_address_distance(121.543462, 25.02612)
#print(f"兩點距離為：{str(instance)}公里")
```

```python
#列出該地點與任一車站的最短距離以及該車站的級別
def short_dist(city_df, city_train):
  min_dist = 100
  shortest_dist = []
  station_level = []
  for i in range(0,len(city_df)):
    for j in range(0, len(city_train)):
      d = get_address_distance(city_train['lng'][j], city_train['lat'][j], city_df['經度'][i], city_df['緯度'][i])
      #print(d)
      if d<min_dist:
        min_dist = d
        level_flag = j
    shortest_dist.append(min_dist)
    station_level.append(city_train['車站級別'][level_flag])
    min_dist = 100
  city_df['與車站最短距']=shortest_dist
  city_df['車站級別']=station_level
  return city_df
```

```python
def kimen_airplane_dist(city_df):
  city_df['車站級別'] = None
  shortest_dist = []
  for i in range(0,len(city_df)):
    d = get_address_distance(118.3611111, 24.4288889, city_df['經度'][i], city_df['緯度'][i])
    shortest_dist.append(d)
  city_df['與車站最短距'] = d
  city_df['車站級別'] = 5
  return city_df
```

```python
from math import radians, cos, sin, asin, sqrt

#address_me = [120.2640171785886,22.831598506259798]

def get_address_distance(station_lon, station_lat, train_lon,  train_lat):
    lon1, lat1, lon2, lat2 = map(radians, [station_lon, station_lat, train_lon, train_lat])
    # 半正矢公式
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    c = 2 * asin(sqrt(a))
    r = 6371
    dis = round((c * r * 1000)/1000, 2)
    return dis
```

```python
#列出該地點與任一笑的最短距離以及該笑總人數
def school_short_dist_n_people(city_df, city_school, col_name, col_short_dist_name):
  min_dist = 100
  shortest_dist = []
  people = []
  for i in range(0,len(city_df)):
    for j in range(0, len(city_school)):
      d = get_address_distance(city_school['lng'][j], city_school['lat'][j], city_df['經度'][i], city_df['緯度'][i])
      #print(d)
      if d<min_dist:
        min_dist = d
        loc_flag = j
    shortest_dist.append(min_dist)
    people.append(city_school[col_name][loc_flag])
    min_dist = 100
  city_df[col_short_dist_name]=shortest_dist
  city_df[col_name]=people
  return city_df
```

```python
#計算經緯度距離
from math import radians, cos, sin, asin, sqrt

def get_address_distance(station_lon, station_lat, train_lon,  train_lat):
    lon1, lat1, lon2, lat2 = map(radians, [station_lon, station_lat, train_lon, train_lat])
    # 半正矢公式
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    c = 2 * asin(sqrt(a))
    r = 6371
    dis = round((c * r * 1000)/1000, 2)
    return dis
```

```python
#列出該地點與任一捷運站的最短距離
#第一次計算
def short_dist(city_mrt, city_pt):
  min_dist = 1000
  shortest_dist = []
  station_level = []
  for i in range(0,len(city_mrt)):
    for j in range(0, len(city_pt)):
      d = get_address_distance(city_pt['lng'][j], city_pt['lat'][j], city_mrt['經度'][i], city_mrt['緯度'][i])
      #print(d)
      if d<min_dist:
        min_dist = d
        level_flag = j
    shortest_dist.append(min_dist)
    station_level.append(city_pt['級別'][level_flag])
    min_dist = 1000
  city_mrt['與捷運站最短距']=shortest_dist
  city_mrt['捷運級別']=station_level
  return city_mrt
```

```python
#一次以上站點距離計算
def more_check_dist(city_mrt, second_pt):
  min_dist = 1000
  for i in range(0,len(city_mrt)):
    for j in range(0, len(second_pt)):
      d = get_address_distance(second_pt['lng'][j], second_pt['lat'][j], city_mrt['經度'][i], city_mrt['緯度'][i])
      if d < city_mrt['與捷運站最短距'][i]:
        min_dist = d
        level_flag = j
      else:
        min_dist = city_mrt['與捷運站最短距'][i]
        level_flag = None
    city_mrt.loc[i,'與捷運站最短距'] = min_dist
    if level_flag != None:
      city_mrt.loc[i, '捷運級別'] = second_pt['級別'][level_flag]
      #city_mrt['捷運級別'][i] = second_pt['級別'][level_flag]#會改到dataframe副本 出錯
    min_dist = 1000
  return city_mrt
```

```python
from math import radians, cos, sin, asin, sqrt

#address_me = [120.2640171785886,22.831598506259798]

def get_address_distance(station_lon, station_lat, train_lon,  train_lat):
    lon1, lat1, lon2, lat2 = map(radians, [station_lon, station_lat, train_lon, train_lat])
    # 半正矢公式
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    c = 2 * asin(sqrt(a))
    r = 6371
    dis = round((c * r * 1000)/1000, 2)
    return dis
```

```python
#列出該地點與任一笑的最短距離以及該笑總人數
def bus_count(city_df, city_bus, col_bus_count_name):
  count_list = []
  for i in range(0,len(city_df)):
    count=0
    for j in range(0, len(city_bus)):
      d = get_address_distance(city_bus['lng'][j], city_bus['lat'][j], city_df['經度'][i], city_df['緯度'][i])
      #print(d)
      if d<0.5:
        count+=1
    count_list.append(count)
  city_df[col_bus_count_name]=count_list
  return city_df
```

```python
from math import radians, cos, sin, asin, sqrt

#address_me = [120.2640171785886,22.831598506259798]

def get_address_distance(station_lon, station_lat, train_lon,  train_lat):
    lon1, lat1, lon2, lat2 = map(radians, [station_lon, station_lat, train_lon, train_lat])
    # 半正矢公式
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    c = 2 * asin(sqrt(a))
    r = 6371
    dis = round((c * r * 1000)/1000, 2)
    return dis
```

```python

def short_dist(city_df, city_train):
  min_dist = 100
  shortest_dist = []
  for i in range(0,len(city_df)):
    for j in range(0, len(city_train)):
      d = get_address_distance(city_train['lng'][j], city_train['lat'][j], city_df['經度'][i], city_df['緯度'][i])
      #print(d)
      if d<min_dist:
        min_dist = d
        level_flag = j
    shortest_dist.append(min_dist)
    min_dist = 100
  city_df['與百貨最短距']=shortest_dist #百貨or高鐵
  return city_df
```
